# 7° Manipulando arquivos em Banco de Dados SQLite

```
# Acessando dados do Banco de Dados
import pandas as pd
import sqlite3

# open engine connection
con=sqlite3.connect('/content/drive/MyDrive/CDPro - Aula 1/cdpro 05.10/escola.db')

# create a cursor object
cur = con.cursor()
```

```
# Criar uma Consulta: Query
Query = cur.execute('select * from cursos')
```

```
# Save results of the query to DataFrame: df
df = pd.DataFrame(Query.fetchall(), columns=['ID','TI','CA'])

# Close connection
con.commit()

# Print head of DataFrame df
df
```
